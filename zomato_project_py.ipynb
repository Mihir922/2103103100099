```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# New Section

```python
def handlerate(value):
  value = str(value).split("/")
  value = value[0];
  return float(value)

dataframe["rate"] = dataframe["rate"].apply(handlerate)
dataframe["rate"].head(5)
```

```python


# Sample DataFrame (use your actual DataFrame)
dataframe = pd.DataFrame({
    'listed_in(type)': ['Type1', 'Type1', 'Type2', 'Type2'],
    'online_order': ['Yes', 'No', 'Yes', 'No'],
    'count': [100, 200, 150, 250]
})

# Ensure 'count' is numeric
dataframe['count'] = pd.to_numeric(dataframe['count'], errors='coerce')

# Create the pivot table
pivot_table = pd.pivot_table(dataframe,
                             index="listed_in(type)",
                             columns="online_order",
                             values="count",  # Ensure you are summarizing a numeric column
                             aggfunc="sum")

# Calculate percentages
pivot_table_percent = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100

# Plot the heatmap
plt.figure(figsize=(5,3))
sns.heatmap(pivot_table_percent, annot=True, cmap="YlGnBu", fmt=".1f")
plt.title("Heatmap (Percentages)")
plt.show()
```
